```python
import matplotlib.pyplot as plt
```

```python
# Example data
year = [1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020]
pop = [2.5, 3.0, 3.7, 4.4, 5.3, 6.1, 6.9, 7.8]

# Print the last item from year and pop
print(year[-1])  # Should print 2020
print(pop[-1])   # Should print 7.8

# Make a line plot: year on the x-axis, pop on the y-axis
plt.plot(year, pop)

# Display the plot with plt.show()
plt.show()
```

```python
import numpy as np

# Example data
gdp_cap = [974.5803384, 5937.029526, 6223.367465, 4797.231267, 12779.37964, 34435.36744, 36126.4927]
life_exp = [43.828, 76.423, 72.301, 42.731, 75.32, 81.235, 79.829]
pop = [31.889923, 3.600523, 33.333216, 12.420476, 40.301927, 20.434176, 8.199783]  # Population in millions

# Convert pop to a NumPy array and double the values
np_pop = np.array(pop) * 2

# Define labels and title
xlab = 'GDP per Capita [in USD]'
ylab = 'Life Expectancy [in years]'
title = 'World Development in 2007'

# Tick values and labels for x-axis
tick_val = [1000, 10000, 100000]
tick_lab = ['1k', '10k', '100k']

# Create a scatter plot with sizes based on population
plt.scatter(gdp_cap, life_exp, s=np_pop)

# Customize the plot with labels, title, and ticks
plt.xlabel(xlab)
plt.ylabel(ylab)
plt.title(title)
plt.xticks(tick_val, tick_lab)

# Display the plot
plt.show()
```

```python
#import matplotlib.pyplot as plt
import numpy as np

# Example data
gdp_cap = [974.5803384, 5937.029526, 6223.367465, 4797.231267, 12779.37964, 34435.36744, 36126.4927]
life_exp = [43.828, 76.423, 72.301, 42.731, 75.32, 81.235, 79.829]
pop = [31.889923, 3.600523, 33.333216, 12.420476, 40.301927, 20.434176, 8.199783]  # Population in millions

# Correct length of colors list to match data points
col = ['red', 'green', 'blue', 'yellow', 'black', 'cyan', 'magenta']  # Color for each data point

# Convert pop to a NumPy array and double the values
np_pop = np.array(pop) * 2

# Define labels and title
xlab = 'GDP per Capita [in USD]'
ylab = 'Life Expectancy [in years]'
title = 'World Development in 2007'

# Tick values and labels for x-axis
tick_val = [1000, 10000, 100000]
tick_lab = ['1k', '10k', '100k']

# Create a scatter plot with sizes and colors based on population and continent
plt.scatter(gdp_cap, life_exp, s=np_pop, c=col, alpha=0.8)

# Customize the plot with labels, title, and ticks
plt.xlabel(xlab)
plt.ylabel(ylab)
plt.title(title)
plt.xticks(tick_val, tick_lab)

# Display the plot
plt.show()
```

```python
# Example data
gdp_cap = [974.5803384, 5937.029526, 6223.367465, 4797.231267, 12779.37964, 34435.36744, 36126.4927]
life_exp = [43.828, 76.423, 72.301, 42.731, 75.32, 81.235, 79.829]
pop = [31.889923, 3.600523, 33.333216, 12.420476, 40.301927, 20.434176, 8.199783]  # Population in millions
col = ['red', 'green', 'blue', 'yellow', 'black', 'cyan', 'magenta']  # Color for each data point

# Convert pop to a NumPy array and double the values
np_pop = np.array(pop) * 2

# Define labels and title
xlab = 'GDP per Capita [in USD]'
ylab = 'Life Expectancy [in years]'
title = 'World Development in 2007'

# Tick values and labels for x-axis
tick_val = [1000, 10000, 100000]
tick_lab = ['1k', '10k', '100k']

# Create a scatter plot with sizes and colors based on population and continent
plt.scatter(gdp_cap, life_exp, s=np_pop, c=col, alpha=0.8)

# Add text annotations
plt.text(5000, 70, 'India', fontsize=20)
plt.text(12000, 75, 'China', fontsize=20)

# Customize the plot with labels, title, ticks, and gridlines
plt.xlabel(xlab)
plt.ylabel(ylab)
plt.title(title)
plt.xticks(tick_val, tick_lab)
plt.grid(True)  # Enable gridlines

# Display the plot
plt.show()
```
